### Loading file with delimiter

```python
df1 = pd.read_csv('transaction.csv',delimiter=';')
df1.head()
```

### Summary of the data

```python
df1.info()
```

### Creating new columns of CostPerTransaction

```python
df1['CostPerTransaction'] = df1['CostPerItem'] * df1['NumberOfItemsPurchased']
```

```python
df1.info()
```

```python
df1.head()
```

### Creating SalesPerTransaction Column

```python
df1['SalesPerTransaction'] = df1['SellingPricePerItem'] * df1['NumberOfItemsPurchased']
```

```python
df1.head()
```

### Calculation of Profit and Markup

```python
df1['ProfitPerTransaction'] = df1['SalesPerTransaction'] - df1['CostPerTransaction']
```

```python
df1.head()
```

```python
df1['Markup'] = (df1['SalesPerTransaction'] - df1['CostPerTransaction']) / df1['CostPerTransaction']
```

```python
df1.head()
```

```python
df1.info()
```

### Decimal correction using round() function

```python
df1['Markup'] = df1['Markup'].round(2)
```

```python
df1.head()
```

### Concatenating Day, Month and Year Columns

```python
df1['Date'] = df1['Day'].astype(str) + '-' + df1['Month']+ '-' + df1['Year'].astype(str)
```

```python
df1.Date.head()
```

### Split function in ClientKeyWords Column

```python
split_cols = df1['ClientKeywords'].str.split(',',expand=True)
split_cols
```

```python
df1['ClientAge'] = split_cols[0]
df1['ClientType'] = split_cols[1]
df1['LengthOfContract'] = split_cols[2]
```

```python
df1.head()
```

### Replace function to replace necessary elements

```python
df1['ClientAge'] = df1['ClientAge'].str.replace('[','')
df1['LengthOfContract'] = df1['LengthOfContract'].str.replace(']','')
```

```python
df1.head()
```

### Lower functions for Item Description

```python
df1['ItemDescription'] = df1['ItemDescription'].str.lower()
```

```python
df1.head()
```

### Merging season dataset into the main dataset

```python
df2 = pd.read_csv("value_inc_seasons.csv",delimiter=';')
df2.head()
```

```python
df3 = pd.merge(df1,df2,on='Month')
df3.head()
```

### Drop unwanted columns

```python
df3 = df3.drop(['Day','Month','Year','ClienttKeywords'],axis=1)
df3.head()
```

### Export to new csv file

```python
df3.to_csv('value_inc_cleaned.csv',index=False)
```
